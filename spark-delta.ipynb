```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("DeltaExample")
    .config("spark.jars.packages", "io.delta:delta-spark_2.12:3.2.0")
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension")
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")
    .getOrCreate()
)

print("Extensões ativas:", spark.conf.get("spark.sql.extensions"))
print("Catálogo ativo:", spark.conf.get("spark.sql.catalog.spark_catalog"))
print("Extensões ativas:", spark.conf.get("spark.sql.extensions"))
print("Catálogo ativo:", spark.conf.get("spark.sql.catalog.spark_catalog"))

spark.sql("CREATE DATABASE IF NOT EXISTS filmes_delta")
spark.sql("USE filmes_delta")

spark.sql("""
    CREATE TABLE IF NOT EXISTS top_filmes (
        titulo STRING,
        ano INT,
        nota DOUBLE
    )
    USING delta
""")

print("Tabela Delta criada.")

csv_path = "top_100_movies_full_best_effort.csv"
df = spark.read.option("header", True).csv(csv_path)
print("Dataset carregado!")
df.show(5)

df.write.format("delta").mode("overwrite").saveAsTable("filmes_delta.top_filmes")
print("Tabela Delta populada.")

print("Exemplo SELECT:")
spark.sql("SELECT * FROM filmes_delta.top_filmes LIMIT 5").show()

print("Exemplo INSERT:")
spark.sql("INSERT INTO filmes_delta.top_filmes VALUES ('Novo Filme', 2025, 9.8)")

print("Exemplo UPDATE:")
spark.sql("UPDATE filmes_delta.top_filmes SET nota = 9.9 WHERE titulo = 'Novo Filme'")

print("Exemplo DELETE:")
spark.sql("DELETE FROM filmes_delta.top_filmes WHERE ano < 2000")

print("Operações finalizadas!")

spark.stop()
```
